```python
# load the basic informatin of the cars
df = pd.read_pickle('basic_data_merged.pkl')
print df
df.head()
```

```python
# extract actual model from the title
possible_models = set(df.model)
possible_models = [s.lower() for s in possible_models]

title = [df.title, df.model]
model = []
for k, t in enumerate(title[0]):
    try:
        t = t.split(":")[0]
        if t.lower() not in possible_models:
            t = title[1][k]
    except Exception, e:
        print e
        t = title[1][k]
    model.append(t)
# no exception printed. So we are good!

df['true_model'] = model
    
```

```python
# Process the columns to change datatype 
duration = [int(s.replace("days", "").replace("day","")) for s in df.duration]
df['duration'] = duration
df.head()


# change price
price = [float("".join(re.findall(r'[0-9]|\.',s))) for s in df.price]
df.price = price
df['prod_milage'] = df.prod_milage.astype(float)
df['prod_yr'] = df.prod_yr.astype(int)
df.dtypes
```

```python
fake_df = df[['true_model']]
count_by_model = fake_df.groupby('true_model').count().sort('true_model', ascending=False)
count_by_model.head(10)
```

```python
rcParams['font.size'] = 11
labels = list(count_by_model.index[:20])
labels.append('Other')
fracs = list(count_by_model.true_model[:20])
fracs.append(np.sum(count_by_model.true_model[20:]))
pickle.dump([labels, fracs], open('f.pkl','wb'))
fig = plt.figure(1, figsize=(10,10))
plt.axes([0.1, 0.1, 0.8, 0.8])

plt.pie(fracs, labels=labels,
                autopct='%1.1f%%', startangle=0, radius= 1.2)
# plt.title('Percentage breakdown for different card models')
print 'Percentage breakdown for different car models'
# ax.patch.se
# plt.savefig('pie.png', dpi=300)
plt.show()
rcParams['font.size'] = 14
```

```python
# price analysis
price_df = df[['true_model','price']]
price_df = price_df[price_df.price > 100]
price_df = price_df.groupby('true_model').mean().sort('price', ascending=False)
price_df.head()
plt.hist(price_df.price, bins=20)
plt.xlabel('Avergae Price (USD)')
plt.ylabel('Number of models')
plt.title('Average price vs models')
plt.show()
```

```python
# see when most people bid
fin_ts = 'time_series_big.csv'
df = pd.read_csv(fin_ts, sep='\t' , names=\
		['auto_flag', 'price', 'bid_time', 'user', 'prod_id'])
# format = "%b-%d-%y %H:%M:%S"
# bid_times = [time.strptime(s[:-4], format) for s in df.bid_time]

bid_times = []
for s in df.bid_time:
    try:
        s = s.split(" ")[4].split(":")
        bid_times.append(s)
    except Exception, e:
        print e
        
# bid_times = [s.split(" ") for s in df.bid_time[:20] ]
# print bid_times

arr = np.array([1.0, 1/60.0, 1/3600.0])
bid_hours = [np.dot(np.array(s, dtype=np.float),arr) for s in bid_times ]

plt.hist(bid_hours, bins = np.array(range(25)))

plt.xlabel('Hours (24 hour clock)')
plt.ylabel('Number of bids')
plt.title('When most people bid')
plt.savefig("bid_time.png")
plt.show()
```

```python
# distribution of number of bids and # bidders
n_bids = df.num_bids
plt.hist(n_bids, bins = 30)
plt.xlabel('Number of products')
plt.ylabel('Number of bids')
plt.title('Bid distribution')
# plt.savefig("bid_time.png")
plt.show()
```

```python
n_bidders = df.num_bidder
plt.hist(n_bidders, bins = 30, normed=True)
plt.xlabel('Number of products')
plt.ylabel('Number of bidders')
plt.title('# bidder distribution')
# plt.savefig("bid_time.png")
plt.show()
```
